# Getting data

## Importing data


```python
train_dataset = pd.read_csv("data/beans_train.csv")
test_dataset = pd.read_csv("data/test_public.csv")
```

```python
train_dataset
```

## Formatting data

```python
train_np = train_dataset.to_numpy()
test_np = test_dataset.to_numpy()
```

```python
classes = np.unique(train_np[:,-1])
classes
```

```python
classes_dic = {class_name:i for i,class_name in enumerate(classes)}
classes_dic
```

```python
x_train = train_np[:, :-1]
y_train = np.vectorize(classes_dic.get)(train_np[:,-1])
```

```python
x_test = test_np
```

# Simple models

## Logistic regression (pas bon)

```python
hyperParams_LR = {'solver':['newton-cg', 'sag', 'saga', 'lbfgs'],
                       'penalty':['l1','l2', 'elasticnet'],
                       'class_weight': [None, 'balanced'],
                       #'max_iter' : [3000],
                       'C' : np.linspace(0.01,0.2,12),
                       #'C' : np.linspace(0.11,0.12,10),
                       'solver':['saga'],
                       'multi_class': ['multinomial']
                       }

clf_LR_gridSearch =GridSearchCV(estimator=LogisticRegression(),param_grid=hyperParams_LR, scoring='f1_macro', verbose=5)

clf_LR_gridSearch.fit(x_train,y_train)
best_prams_LR = clf_LR_gridSearch.best_params_
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
```

```python
clf = RandomForestClassifier(random_state=10, n_estimators=100, ccp_alpha=4e-5)
clf.fit(X_train1,y_train1)
y_pred1 = clf.predict(X_test1)
f1_score(y_test1, y_pred1, average="macro")
```

```python

def translate_numerical_prediction(predictions, classes_dic):
    inverse_class_dic = {n:c for c,n in classes_dic.items()}
    return np.vectorize(inverse_class_dic.get)(predictions)

print(translate_numerical_prediction(y_pred1, classes_dic))
```

```python
clf_complete = RandomForestClassifier(random_state=10, n_estimators=100, ccp_alpha=4e-5)
clf_complete.fit(x_train,y_train)
```

```python
y_pred = clf_complete.predict(x_test)
y_pred_cat = translate_numerical_prediction(y_pred, classes_dic)
```

```python
def make_submission_file(predictions, classes_dic, subfilename):
    pred_cat = translate_numerical_prediction(predictions, classes_dic)
    df = pd.DataFrame()
    df["id"] = np.arange(1, len(pred_cat)+1)
    df["class"] = pred_cat
    df.to_csv(subfilename, index=False)
```
